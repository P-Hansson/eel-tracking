```python
#### Create a video file with only every 5th frame

import cv2
import sys
import os
import shutil

def clear_filepath(directory):
    for dirpath,_, filenames in os.walk(directory):
        for files in filenames:
            if not files.startswith("."):
                yield os.path.abspath(os.path.join(dirpath,files))

video_list_path = list(clear_filepath("eel_detection/videos/all_frames"))
print(video_list_path)

for video in video_list_path:
    out_filename = os.path.basename(video)
    print(out_filename)
    dest_in = ("eel_detection/videos/all_frames_done/" + out_filename)
    
    #Clears up the filenames - replace any wrong or problematic symbols with "_" and remove "_trim"
    out_filename = out_filename.replace(" ", "_")
    out_filename = out_filename.replace("-", "_")
    out_filename = out_filename.replace("_trim", "")
    dest_out = ("eel_detection/videos/missing_boundaries/" + out_filename)
    
    in_vid = cv2.VideoCapture(video)
    
    if not in_vid.isOpened():
        print("Cannot open input video file")
        sys.exit()
    
    ok, frame = in_vid.read()
    if not ok:
        print("Cannot read video file")
        sys.exit
    
    width, height = frame.shape[1], frame.shape[0]
    
    fps = in_vid.get(cv2.CAP_PROP_FPS)
    
    out_vid = cv2.VideoWriter(out_filename, cv2.VideoWriter_fourcc(*"XVID"), fps, (width, height))
    
    frame_counter = 1
    
    while True:
        # To change the number of frames saved, change the number in the following row - 
        # it is currently set to 5, meaning the code saves every fifth frame
        if ((frame_counter % 5) == 0):
            out_vid.write(frame)
            
        ok, frame = in_vid.read()
        if not ok:
            break
            
        frame_counter += 1
        
    out_vid.release()
    shutil.move(out_filename, dest_out)
    
    in_vid.release()
    shutil.move(video, dest_in)

print("Done")
```
